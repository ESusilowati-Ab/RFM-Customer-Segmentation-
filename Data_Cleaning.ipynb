```python
import pandas as pd

# Reading main files
orders = pd.read_csv('olist_orders_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')
customers = pd.read_csv('olist_customers_dataset.csv')
translation = pd.read_csv('product_category_name_translation.csv')
products = pd.read_csv('olist_products_dataset.csv')
```

```python
# Merge order data with purchased items
df_orders = pd.merge(orders, items, on='order_id', how='inner')
```

```python
# Added customer location data
df_complete = pd.merge(df_orders, customers, on='customer_id', how='inner')
```

```python
# Combine product info with language translation
products_translated = pd.merge(products, translation, on='product_category_name', how='left')

# Merge translated product data into main data using product_id
df_final = pd.merge(df_complete, products_translated, on='product_id', how='left')
```

```python
# Final Merge to Master Data
df_final.head()
```

```python
df_final.to_csv('olist_master_dataset.csv', index=False)
```

```python
print(f"Number of rows before final merge: {df_complete.shape[0]}")
print(f"Number of rows after final merge: {df_final.shape[0]}")
```

```python
df_final['product_category_name_english'] = df_final['product_category_name_english'].fillna('unknown')
```

```python
datetime_cols = ['order_purchase_timestamp', 'order_approved_at', 
                 'order_delivered_carrier_date', 'order_delivered_customer_date', 
                 'order_estimated_delivery_date']

for col in datetime_cols:
    df_final[col] = pd.to_datetime(df_final[col])
```
